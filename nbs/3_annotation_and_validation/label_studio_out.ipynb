```python
#import modules


from pathlib import Path

import os

import sys

sys.path.append(str(Path("../../src").resolve()))

import importlib
import json_formatting

importlib.reload(json_formatting)

from json_formatting import Formatter
```

## Convert from json format to jsonl

```python
# set paths to each of the annotaters output paths

markus_label_studio = Path(os.path.join("..",
                           "..",
                           "data",
                           "training_data",
                           "validation_set",
                           "intermediate_files",
                           "markus_annotated.json"))

rune_label_studio = Path(os.path.join("..",
                           "..",
                           "data",
                           "training_data",
                           "validation_set",
                           "intermediate_files",
                           "rune_annotated.json"))

markus_LS_jsonl = Path(os.path.join(os.path.dirname(markus_label_studio),
                                    "markus_gold_annotated.jsonl"))

rune_LS_jsonl = Path(os.path.join(os.path.dirname(rune_label_studio),
                                    "rune_gold_annotated.jsonl"))


#set path to save merged by inter annotater agreeement
final_validation_outpath = Path(os.path.join(os.path.dirname(markus_label_studio),
                           "..",
                           "validation_set.jsonl"))
```

### Reformat from json to jsonl

```python
json_input_files = [markus_label_studio, rune_label_studio]
jsonl_output_files = [markus_LS_jsonl, rune_LS_jsonl]

FM = Formatter()

for inp, oup in zip(json_input_files, jsonl_output_files):

    FM.json_to_jsonl(inp,oup)
```

### Only keep sentences with agreement between annotaters for final validation. Also compute statistics

```python
FM.compare_LS_files(markus_LS_jsonl, 
                    rune_LS_jsonl, 
                    output_file = final_validation_outpath)
```
