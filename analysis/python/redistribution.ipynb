```python
import pandas as pd
```

```python
path = "./1999-2024_CHES_dataset_meansV2.csv"

data=pd.read_csv(path)
data = data.loc[data["country"]==2]
```

```python
parties_to_keep = ["S", "DF", "EL", "SF", "V", "RV", "KD", "LA", "DD", "KF", "ALT", "NB", "M" ]
```

```python
df_sub = data[["party", "lrgen"]].dropna()
df_sub["party"] = df_sub["party"].str.upper()
df_sub = df_sub.replace("KRF", "KD")
df_sub = df_sub.replace("NY", "LA")

df_sub = df_sub[df_sub["party"].isin(parties_to_keep)]

df_sub["lrgen_z"] = (df_sub["lrgen"] - df_sub["lrgen"].mean()) / df_sub["lrgen"].std()

df_party = (
    df_sub.groupby("party", as_index=False)
    .agg(mean_lrgen=("lrgen_z", "mean"))
)

df_party["bloc"] = df_party["mean_lrgen"].apply(
    lambda x: "left_bloc" if x > 0 else "right_bloc"
)

df_party["wingness"] = df_party["mean_lrgen"].abs()
df_party.sort_values(by="mean_lrgen", ascending=True)
```

```python
df_party.to_csv("./bloc_and_wingness.csv")
```
