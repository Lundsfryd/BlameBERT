Power analysis to estimate validation set size

```python
from statsmodels.stats.proportion import proportion_effectsize
from statsmodels.stats.power import NormalIndPower, zt_ind_solve_power
import numpy as np

def validation_set_size(p_expected, delta, alpha=0.05, power=0.80):
    """
    p_expected: pilot accuracy estimate
    delta:      minimum detectable difference
    """
    p_null = p_expected - delta
    effect_size = proportion_effectsize(p_expected, p_null)

    analysis = NormalIndPower()
    n = analysis.solve_power(
        effect_size=abs(effect_size),
        alpha=alpha,
        power=power,
        alternative='two-sided'
    )
    return int(np.ceil(n))
```

# If calculating based on accuracy

```python
n = validation_set_size(p_expected=0.85, delta=0.05, alpha=0.05, power=0.80)

print(f"Recommended validations set size of {n}")
```

# If calculating based on recall

```python
n = validation_set_size(p_expected=0.72, delta=0.05, alpha=0.05, power=0.80)
print(f"Recommended validations set size of {n}")
```

# Might be more correct to do one sided

```python
from statsmodels.stats.power import zt_ind_solve_power
from statsmodels.stats.proportion import proportion_effectsize
from statsmodels.stats.power import NormalIndPower
import numpy as np

def validation_set_size_onesample(p_expected, delta, alpha=0.05, power=0.80):
    """
    One-sample proportion test.
    p_expected: observed pilot accuracy/recall
    delta:      minimum detectable difference
    """
    p_null = p_expected - delta  # baseline to detect against
    effect_size = proportion_effectsize(p_expected, p_null)  # Cohen's h

    from statsmodels.stats.power import GofChisquarePower, zt_ind_solve_power
    # One-sample z-test for proportion
    from statsmodels.stats.power import NormalIndPower
    n = zt_ind_solve_power(
        effect_size=abs(effect_size),
        alpha=alpha,
        power=power,
        alternative='two-sided',
        nobs1=None,
        ratio=0  # ratio=0 means one-sample
    )
    return int(np.ceil(n))
```

# Based on accuracy

```python
n = validation_set_size_onesample(p_expected=0.85, delta=0.05, alpha=0.05, power=0.80)

print(f"Recommended validations set size of {n}")
```
