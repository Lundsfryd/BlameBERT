# Machine tranlsation

This notebook desribes the pipeline for machine translation of the sentences to be used for modelling. Machine translation is nessecary due to prelimenaryt blame detection using the DEBATE model.

The following steps will be implemented:

- Read in modelling data (.jsonl)
- Apply Machine translation using the OPUS MT pipeline
- Save as .jsonl to be passed to prelimenary blame detection

Note to self: use environment blameBERT python version 3.12

```python
import os
from pathlib import Path

from machine_trans import MachineTranslation

inpath = Path(os.path.join("..",
                      "..",
                      "raw_data",
                      "test_json_modelling_data.jsonl"))

output_path = Path(os.path.join("..",
                      "..",
                      "raw_data",
                      "test_json_modelling_data_translated.jsonl"))
```

```python
MT = MachineTranslation(jsonl_path = inpath, 
                        outpath = output_path, 
                        batch_size=2)

MT.run_translation()
```

The above translation pipeline can also be run from terminal as specified beneith

python machine_trans.py --input_path_jsonl <> --output_path_jsonl <> 

optional parameters
--batch_size (default: 2)
